# Detecting Multicolinearity in Data:

```python
def multicoliniarity(data,threshold):
    output=set()
    corr=data.corr()
    for i in range(len(corr.columns)):
        for j in range(i):
            if abs(corr.iloc[i,j])>threshold:
                b=corr.columns[i]
                output.add(b)
                
    return(output)
```

```python
multicol=multicoliniarity(x_data,0.90)
multicol
```

```python
len(multicol)
```

# Splitting the dataset into Dependent and Independent columns(train and test):

```python
from sklearn.model_selection import train_test_split
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(data,target_data,train_size=.75)
```

```python
xtrain
```

```python
ytrain
```

# Classification Algorithms:

```python
from sklearn.linear_model import LogisticRegression
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
from sklearn.svm import SVC
```

```python
from sklearn.naive_bayes import GaussianNB
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
from xgboost import XGBClassifier
```

```python
from sklearn.ensemble import GradientBoostingClassifier
```

```python
result=[]
```

```python
models={'log':{'model':LogisticRegression()},
        'dtree':{'model':DecisionTreeClassifier()},
        'svc':{'model':SVC()},
        'gnb':{'model':GaussianNB()},
       'rforest':{'model':RandomForestClassifier(n_estimators=500)},
       'xgb':{'model':XGBClassifier()},
       'gboost':{'model':GradientBoostingClassifier()}}
```

```python
for i in models:
    a=models.get(i)
    model=a.get('model')
    model.fit(xtrain,ytrain)
    score=model.score(xtest,ytest)
    out={'model':i,'score':score}
    result.append(out)
```

```python
final=pd.DataFrame(result)
final
```

```python
plt.barh(final.model,final.score)
plt.title('Accuracy of different Classification algorithms')
plt.xlabel('Count in %')
plt.grid()
```

```python
# Considering the different models, XGB Classifier gives more accurancy
```
